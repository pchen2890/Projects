```python
import pandas as pd
```

```python
df = pd.DataFrame.from_csv('hulu.csv',encoding = "ISO-8859-1",index_col=None)
```

```python
df.columns
```

```python
df = df[['show_canonical_name', 'show_clips_count', 'show_episodes_count', 'show_genre', 'show_genres', 'show_rating', 'show_seasons_count',
       'show_videos_count','show_company_name']]
```

```python
df.columns = ['show_name', 'show_clips_count',
       'show_episodes_count', 'main_genre', 'show_genres', 'show_rating',
       'show_seasons_count', 'show_videos_count', 'show_company_name']
```

```python
df.show_genres = df.show_genres.str.split("~")
```

```python
df_genres = df.show_genres.str.join('|').str.get_dummies()
```

```python
df2 = pd.concat([df,df_genres],axis=1).drop('show_genres',axis=1)
```

```python
# dashboard to get average rating based on criteria, number in genre,  
df2
```

```python
pd.DataFrame.to_csv(df2,'hulu_clean.csv')
```
